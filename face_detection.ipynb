```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('C:\computer vision\opencv\image.jpg')
plt.imshow(img)
```

```python
orginal=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(orginal)
```

```python
face_casecade=cv2.CascadeClassifier('C:\computer vision\opencv\haarcascade_frontalface_default.xml')
```

```python
face_casecade
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray
```

```python
face=face_casecade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=4)
```

```python
face
```

```python
for (x,y,w,h) in face:
    cv2.rectangle(orginal,(x,y),(x+w,y+h),color=(0,255,0),thickness=5)
```

```python
plt.imshow(orginal)
```

```python
video=cv2.VideoCapture(0)
```

```python
while True:
    succes,frame =video.read()

    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_casecade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=5)
    print(len(faces))
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),3)

    
    cv2.imshow('webcam',frame)
    if cv2.waitKey(1) & 0xFF ==27:
        break
video.release()
cv2.destroyAllWindows()
```
